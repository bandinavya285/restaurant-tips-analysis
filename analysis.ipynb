```python
import pandas as pd

# Step 1: Load the dataset using a reliable web URL
url = 'https://raw.githubusercontent.com/pandas-dev/pandas/main/pandas/tests/io/data/csv/tips.csv'
df = pd.read_csv(url)

# Step 2: Inspect data structure (prints information in the terminal window below)
print("--- Dataset Structural Info ---")
df.info()

# Step 3: Clean the data by dropping any missing values
df = df.dropna()

# Display the first 5 rows in a clean table right inside VS Code
df.head()
```

```python
# Step 4: Filter for rows where the table size is greater than 3 people
large_groups_df = df[df['size'] > 3]

# Display the filtered rows
large_groups_df.head()
```

```python
# Step 5: Group by the 'day' column and find the mathematical mean (average) of the 'tip' column
daily_tip_average = df.groupby('day')['tip'].mean()

print("📊 --- Meaningful Insights: Average Tip by Day --- 📊")
print(daily_tip_average)
```
